```python
europe = pd.read_csv('./gapminder_gdp_europe.csv', index_col='country')
europe
```

```python
from IPython.core.pylabtools import figsize
figsize(10, 8)

years = europe.columns.str.strip('gdpPercap_')
years_int = years.astype(int)
europe.columns = years_int
```

```python
europe.min().plot(label='min')
europe.max().plot(label='max')
plt.legend(loc='best')
plt.xticks(rotation=90)
```

```python
europe.min().plot(label='min', x=europe.columns.values)
europe.max().plot(label='max')
plt.legend(loc='best')
plt.xticks(rotation=90)
```
